```python
# 2. A) ¿Cuantos viven en zona urbana o rural?
df2 = (df >>
    select(X.sex, X.address,X.absences))
df2
```

```python
# 3. ¿Cuantos que vivne en zona rural tiene mas de 5 fallas?
df4 = df3[df3['absences'] > 5]
df4.count() # Hay 38 estudiantes que viven en zona rural y tiene mas de 5 fallas
```

```python
#4. Media de las notas de los esrudiantes de colegios Gabtiel Pereira Mosuuinho Silveira
df5 = (df >>
    select(X.school, X.internet,X.G3))
df5
```

```python
df6 = (df5 >>
  group_by(X.school,X.internet) >>
  summarize(media_G3 = X.G3.mean())
)
df6

# Se observa que los estudiantes del colegio Gabriel Pereira que no cuentan con 
# internet tienen una nota menor en comparación a los que sí cuentan con internet en el dicho colegio. 
# En el colegio Mousinho da Silveira ocurre algo similar en los estudiantes que cuentan y que no cuentan con internet, 
# Ambos colegios parecen no definir en el rendimiento promedio del tercer periodo académico. 
```
