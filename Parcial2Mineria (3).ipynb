```python
# Crear Dummy
Studytime_dum = pd.get_dummies(Base['studytime'])
Studytime_dum = Studytime_dum.loc[:,['2','3','4']] # ref. 1

failures_dum = pd.get_dummies(Base['failures'])
failures_dum = failures_dum.loc[:,['1','2','3']] # ref. 0

famsup_dum = pd.get_dummies(Base['famsup'])
famsup_dum = famsup_dum.loc[:,['yes']] # ref. no

higher_dum = pd.get_dummies(Base['higher'])
higher_dum = higher_dum.loc[:,['yes']] # ref. no

internet_dum = pd.get_dummies(Base['internet'])
internet_dum = internet_dum.loc[:,['yes']] # ref. no
```

```python
Studytime_dum = Studytime_dum.rename(columns = { '2':'study2',
                                 '3':'study3' ,
                                 '4':'study4'})
failures_dum = failures_dum.rename(columns = { '1':'fail1',
                                 '2':'fail2' ,
                                 '3':'fail3'})
famsup_dum = famsup_dum.rename(columns = { 'yes':'famyes'})

higher_dum = higher_dum.rename(columns = { 'yes':'higyes'})

internet_dum =  internet_dum.rename(columns = { 'yes':'intyes'})
```

Modelo 1

```python
# Crear variables para el modelo
y = Base['paid'].values
X = pd.concat([Studytime_dum,failures_dum,famsup_dum,higher_dum,internet_dum,Base['G2']], axis = 1,)
```

```python
# grillita
from itertools import combinations
Parametros = ['l1','l2','none',0.3,0.6,0.9]
Pares = list(combinations(Parametros, 2))

Tabla = pd.DataFrame(Pares, columns = ['penalty','C'])
Tabla
grillita = (Tabla >> 
      mask(Tabla['penalty'] != 0.3, Tabla['penalty'] != 0.6, Tabla['C'] != 'l2',Tabla['C'] != 'none'))
grillita['solver'] = ['liblinear','liblinear','liblinear','lbfgs','lbfgs','lbfgs','lbfgs','lbfgs','lbfgs']
grillita
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
```

```python
Preci_mod1 = []
Exact_mod1 = []
y = y.astype('int')
for i in range(9):
 Log1 = LogisticRegression(C= grillita.iloc[i,1],solver = grillita.iloc[i,2],penalty = grillita.iloc[i,0])
 Log1.fit(X, y)
 y_pred = Log1.predict(X)
 a =  metrics.precision_score(y, y_pred)
 b = metrics.accuracy_score(y, y_pred)
 Preci_mod1.append(a)
 Exact_mod1.append(b)
```

```python
grillita['Preci_mod1'] = Preci_mod1
grillita['Exact_mod1'] = Exact_mod1
grillita
```
